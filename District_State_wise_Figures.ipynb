```

import requests

url_state_district = 'https://api.covid19india.org/v2/state_district_wise.json'
#'https://api.covid19india.org/state_district_wise.json'  #state-district wise

JSONContent = requests.get(url_state_district).json()
type(JSONContent)

```

```
#JSONContent.keys()

#len(JSONContent)
state_district_data = json_normalize(JSONContent)
#state_district_data = pd.DataFrame.from_dict(pd.io.json.json_normalize(df1), orient='columns')
state_district_data
```

```
state_district_data = json_normalize(JSONContent,record_path='districtData',meta='state')
```

```
print(state_district_data.shape)
state_district_data.head(10)
```

```
state_district_data.describe()
```

```
state_district_data['mortality'] = state_district_data['deceased']*100 / state_district_data['confirmed']
state_district_data['closed'] = state_district_data['deceased'] + state_district_data['recovered']
state_district_data['closed_mortality'] = state_district_data['deceased']*100 / state_district_data['closed']
```

```
state_district_data[state_district_data.mortality>0]
```

```
#df.groupby(['Name', 'Fruit'])['Number'].agg('sum') 
state_confirmed_agg = state_district_data.groupby(['state'])['confirmed'].agg('sum').reset_index()
state_confirmed_agg = state_confirmed_agg.rename(columns={'confirmed':'state_confirmed'})
```

```
state_deceased_agg = state_district_data.groupby(['state'])['deceased'].agg('sum').reset_index()
state_deceased_agg = state_deceased_agg.rename(columns={'deceased':'state_deceased'})
```

```
state_recovered_agg = state_district_data.groupby(['state'])['recovered'].agg('sum').reset_index()
state_recovered_agg = state_recovered_agg.rename(columns={'recovered':'state_recovered'})
```

```
#state_district_data.join(state_deceased_agg,how='left',on='state')
state_district_data = pd.merge(state_district_data,state_confirmed_agg[['state','state_confirmed']],on='state')
state_district_data = pd.merge(state_district_data,state_deceased_agg[['state','state_deceased']],on='state')
state_district_data = pd.merge(state_district_data,state_recovered_agg[['state','state_recovered']],on='state')
state_district_data.head(15)
print(state_district_data.columns)
```

```
state_district_data['state_mortality'] = state_district_data['state_deceased']*100 / state_district_data['state_confirmed']

state_district_data['closed_state'] = state_district_data['state_deceased'] + state_district_data['state_recovered']
state_district_data['closed_state_mortality'] = state_district_data['state_deceased']*100 / state_district_data['closed_state']
```

```
state_district_data = state_district_data.reindex(columns=['state', 
                                                           'state_confirmed', 
       'state_deceased', 'state_recovered', 'state_mortality', 'closed_state',
       'closed_state_mortality','district', 'active', 'confirmed', 'deceased', 'recovered', 
       'delta.confirmed', 'delta.deceased', 'delta.recovered', 
       'mortality', 'closed', 'closed_mortality', ]) 
# state_district_data.columns
```
